```python
def uni_cat(df, x):

  print("{} -> {}\n\n".format(x,df[x].value_counts()) )
  print( "missing values: {}".format(df[x].isnull().sum()))


  # barplot
  sns.countplot(data=df, x=x)
  plt.show()
  print("\n\n")

  # Piechart

  freq =list(df[x].value_counts().values)
  cat= list(df[x].value_counts().index)

  plt.pie(freq, labels= cat, autopct='%.0f%%')
  plt.show()



def uni_cont(df, x, bins):
  # histogram
  sns.histplot(data=df, x=x, bins=bins);
  plt.show()
  print("\n\n")

  # kdeplot
  sns.kdeplot(data=df, x=x);
  plt.show()
  print("\n skewness= {}\n".format(df[x].skew()))
  print("\n {}\n".format(df[x].describe()))
  print("\n\n")


  # boxplot
  sns.boxplot(data=df, x=x);
  plt.show()
  print("\n\n")










```
